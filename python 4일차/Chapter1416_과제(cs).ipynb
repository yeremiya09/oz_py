파이썬 2.7 버전과 파이썬 3에서 range 차이를 설명해주세요

파이썬 2.7
range: 리스트를 반환합니다. range(5)는 [0, 1, 2, 3, 4]와 같은 리스트를 생성합니다.  
xrange: 메모리 효율적인 이터레이터를 반환합니다. range 대신 사용할 수 있으며, 메모리 사용을 줄이는 데 유리합니다.

range: 이터레이터를 반환합니다. 메모리 효율적이며, range(5)는 range(0, 5)와 같은 객체를 생성합니다. 실제 숫자를 생성할 때까지 계산을 미루기 때문에 메모리 사용이 적습니다.

요약
