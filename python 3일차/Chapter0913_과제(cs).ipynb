[문제1]

list와 tuple의 주된 차이점 은 무엇인가요?

정답:  
list: 변경 가능, 대괄호 [], 데이터 수정 및 추가 가능.  
tuple: 변경 불가능, 소괄호 (), 데이터 수정 불가.   

[문제2]

Python의 set에 대해 설명해 주세요

정답:  
파이썬의 set은 중복 없는 원소의 집합을 표현하는 자료형입니다. 주로 다음과 같은 특징과 기능을 가지고 있습니다:

특징
중복 없음:

set은 중복된 원소를 허용하지 않습니다. 같은 원소가 두 번 이상 포함될 수 없습니다.
순서 없음:

set은 원소의 순서를 유지하지 않습니다. 따라서 인덱싱이나 슬라이싱이 불가능합니다.
변경 가능:

set은 변경 가능한(mutable) 자료형입니다. 원소를 추가하거나 제거할 수 있습니다.


[문제3]
Python의 list는 연결리스트(linked list) 인가요?

정답:  
파이썬의 list: 동적 배열로 구현되어 있으며, 연속 메모리와 인덱스 기반 접근을 지원합니다.  
연결 리스트: 각 요소가 메모리 상에서 비연속적으로 저장되며, 노드와 참조를 사용하는 데이터 구조입니다

[문제4]
딕셔너리 자료형에 대해 설명해주세요

정답 :  
중괄호를 사용하여 생성합니다 키 값 쌍을 저장하는 데이터 구조입니다.  
딕셔너리는 각각의 항목이 키와 값으로 이루어진 쌍입니다. 키는 고유해야 하며, 값은 중복될 수 있습니다. 
딕셔너리는 수정, 추가, 삭제가 가능합니다.  
딕셔너리는 해시 테이블을 사용하여 키를 기반으로 빠른 조회를 제공합니다.
